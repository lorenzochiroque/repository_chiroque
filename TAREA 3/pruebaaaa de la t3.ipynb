La ENAHO recolecta informaci√≥n de estos 34,490 hogares a trav√©s de todos sus m√≥dulos. Para utilizar estos m√≥dulos a la vez, realizaremos una "uni√≥n" (merge).

````python
modulo1_smallr_1 = modulo1_smallr[['CONGLOME', 'VIVIENDA', 'HOGAR']]

modulo341_small_2 = modulo341_small[['CONGLOME', 'VIVIENDA', 'HOGAR']]
````
