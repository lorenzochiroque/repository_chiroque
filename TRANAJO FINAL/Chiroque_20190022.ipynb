```python
import altair as alt
import pandas as pd
```

```python
hpness = pd.read_csv('2019.csv')
hpness
```

```python
 #Para la simplificación de nuestra base de datos, reemplazamos con abreviaciones
nuevos_nombres = {'Freedom to make life choices': 'Freedom', 'Country or region' : 'Country'}
hpness.rename(columns = nuevos_nombres, inplace = True)
hpness.info() #con esto podemos ver todos los elementos de nuestra data y así poder comenzar un posible análisis.
```

```python
hpness[['Score', 'Generosity']].groupby('Score').mean()

#buscamos una posible relación entre Generosity y Freedom
```

Como pudimos observar en la tabla anterior, parece extraño ver cómo es que los valores de generosity suben y bajan. 
Todo esto, independiente a la subida del indice de libertad "score";
por lo cual en el siguiente gráfico trataré de proporcionar la info suficiente para establecer alguna conclusión.

- Así propondré un gráfico ⬇ (únicamente basado en Generosity y Score)
- "Score" ➡ INDICE DE FELICIDAD

```python
#elaboración del gráfico ⬇

alt.Chart(hpness).mark_point().encode(
    x = alt.X("Generosity:Q",  title = "GENEROSIDAD"),
    y = alt.Y("Score:Q" ,  title = "INDICE DE FELICIDAD"),
    color = alt.Color("Generosity", scale=alt.Scale(scheme='purplered')),
    size = alt.Size('Generosity')
    ).interactive().properties(
    title={
      "text": ["Correlación entre el INDICE DE FELICIDAD y GENEROSIDAD"], 
      "subtitle": ["Happiness Report-2019"
                   ],
      "color": "Black",
      "subtitleColor": "Brown"
    })
```

Con esto, podemos afirmar que no hay una gran relación entre la generosidad y el indice de felicidad. 
No se aprecia una ruta o relevancia que indique que mientras más felicidad, más generoso, viceversa.
- Si bien los países con un indice de 7.0 de felicidad son muy generosos, antes de la generosidad 0.25 no se aprecian cambios sustanciales.
- Se aprecia pues, que países que tienen un índice de generosidad de 0.2, puede ser igual o más generoso que alguien con mayor felicidad.

De este modo, buscaré simplificar la correlación entre los diversos items para un trabajo más claro.
- parte de esta simplificación consiste en tomar un nuevo indide de felicidad.
- para este nuevo índice, buscaré sustraer este elemento "Generosity", restarlo con el "Score", y así obtener un nuevo índice de Felicidad
- Eliminar Generosity sin más, obviaría parte del estudio, por lo cual es necesario, efectuar y construir un nuevo SCORE!

```python
hpness['SCORE!'] = hpness.eval('Score - Generosity') #ESTE SCORE! ES EL NUEVO INDICE DE FELICIDAD
hpness.drop(columns = 'Score', inplace = True)
hpness.drop(columns = 'Generosity', inplace = True)
hpness #vemos que la configuración de un nuevo "Score" altera el Overall rank, este ya no confirma cuál país tiene un puntaje (SCORE!) más alto
```

```python
#Como la resta de "Generosity" alteraría los valores del ranking...

hpness.sort_values("SCORE!",ascending = False,  inplace = True) #se necesita este orden del NUEVO INDICE DE FELICIDAD
hpness.drop(columns = 'Overall rank', inplace = True)           #así que eliminamos esa columna(Overall rank).
hpness                                                          #con lo cual obtenemos una nueva lista de los países más felices, siendo 0 el más feliz y 155 el menos feliz.
```

Ahora pues, con este nuevo SCORE!, nos interesa ver cuál es la relación entre el apoyo social y el índice de felicidad; de 
este modo juntamos en un gráfico los siguientes datos:
- SCORE! ➡ NUEVO INDICE DE FELICIDAD 😊
- SOCIAL SUPPORT ➡ APOYO SOCIAL ✊

*APARTE BUSCAREMOS QUÉ POSIBLE RELACIÓN HABRÁ ENTRE LA PERCEPCIÓN DE CORRUPCIÓN Y LA EXPECTATIVA DE VIDA, en conjunto claro*

```python
#elaboración del gráfico ⬇
alt.Chart(hpness).mark_point().encode(
    x = alt.X("SCORE!:Q",  title = "NUEVO INDICE DE FELICIDAD"),
    y = alt.Y("Social support:Q" ,  title = "APOYO SOCIAl"),
    color = alt.Color("Perceptions of corruption", scale=alt.Scale(scheme='lightgreyteal')),
    size = alt.Size('Healthy life expectancy')
    ).interactive().properties(
    title={
      "text": ["Correlación entre LIBERTAD y PBI"], 
      "subtitle": ["Happiness Report-2019"
                   ],
      "color": "Black",
      "subtitleColor": "Brown"
    })
```

Es evidente ahora con el gráfico, la reclación entre el apoyo social (de cualquier fuente), y la felicidad. Ambas son directamente proporcionales. 
- Es así que se esclarece la relación entre felicidad y apoyo social; sin embargo, notamos otros elementos importantes y curiosos

- Si bien, los países más felices son lo que reciben más apoyo social, por otro lado vemos que no hay gran diferencia entre los países que presentan una mayor expectativa de vida

- La expectativa de vida se hace ligeramente mayor cuanta más felicidad o apoyo social presente

- Ahora, un elemento curioso de este gráfico, es que con nitidez se aprecia que los países en donde se percibe mayor corrupción(concentración de azul), justamente son en los más felices o que tienen mayor apoyo social. 

Con esto confirmamos las relaciones directas entre felicidad, apoyo social, orrupción y expectativa de vida; y ahora pues, tocaría analizar el rol de la libertad y el PBI per capita

Ahora, para concluir sería interesante ver la relación entre la libertad y el PBI; de 
este modo juntamos en un gráfico los siguientes datos
- Freedom ➡ LIBERTAD 🗽
- GDP per capita ➡ PBI per capita 💲

```python
#elaboración del gráfico ⬇
alt.Chart(hpness).mark_point().encode(
    x = alt.X("SCORE!:Q",  title = "NUEVO INDICE DE FELICIDAD"),
    y = alt.Y("Freedom:Q" ,  title = "LIBERTAD"),
    color = alt.Color("SCORE!", scale=alt.Scale(scheme='lightgreyred')),
    size = alt.Size('GDP per capita')
    ).interactive().properties(
    title={
      "text": ["Correlación entre LIBERTAD y PBI"], 
      "subtitle": ["Happiness Report-2019"
                   ],
      "color": "Black",
      "subtitleColor": "Brown"
    })
```

Así pues, podemos afirmar que sí hay una gran relación entre LIBERTAD y PBI, así como LIBERTAD y FELICIDAD.
- la concentración de NARANJA (en este caso), en la esquina superior derecha, confirma que los países con mayor PBI son los más felices (INDICE DE FELICIDAD), y en menor magnitud son los que poseen mayor PBI per capita (GDP per capita)
- por otro lado, vemos cómo es que los países más difusos (más blancos), son los menos libres, los menos felices y con menor PBI per capita

Con los 3 gráficos podemos deducir y afirmar varios elementos: 
    - Ya en el 2019 veíamos gran correlación entre la libertad, el apoyo social, y en menor magnitud el PBI, en tanto suma importante al INDICE DE LIBERTAD
    - También vemos que la generosidad no tiene tanta relevancia en este estudio(particularmente en el 2019), por lo cual lo sustraemos del INDICE DE LIBERTAD (formando así el SCORE!)
    - En adición vemos que no hay gran diferencia entre los países que presentan una mayor expectativa de vida. La expectativa de vida se hace ligeramente mayor cuanta más felicidad o apoyo social presente
    - Finalmente obtenemos un dato curioso que es que justamente en los países con mayor apoyo social (y en gran magnitud) se presentan estos mayores casos de percepción de corrupción. Pareciera como si este apoyo social fuera parte importante de los indicios de corrupción.
